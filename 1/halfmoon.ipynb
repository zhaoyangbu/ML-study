```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression

X, y = make_moons(n_samples=1000,noise=0.3,random_state=None)

plt.figure()
plt.plot(X[np.where(y==1)[0], 0], X[np.where(y==1)[0], 1], 'o', c = 'b')
plt.plot(X[np.where(y==0)[0], 0], X[np.where(y==0)[0], 1], 's', c = 'r')

clf = LogisticRegression()
clf.fit(X,y)
i, j = make_moons(n_samples=1000,noise=0.3,random_state=None)

k = clf.predict(i)

plt.figure()
plt.plot(i[np.where(k==1)[0], 0], i[np.where(k==1)[0], 1], 'o', c = 'b')
plt.plot(i[np.where(k==0)[0], 0], i[np.where(k==0)[0], 1], 's', c = 'r')
```
